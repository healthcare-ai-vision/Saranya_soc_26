# Task 2: Medical Imaging Dataset Exploration
This notebook explores two medical imaging datasets and analyzes their properties, class distribution, imbalance, and challenges.


## Dataset 1: NIH Chest X-ray Dataset

### Type of Imaging Data
Chest X-ray images

### Number of Images
112,120 images

### Available Classes / Labels
- Atelectasis
- Effusion
- Infiltration
- Mass
- Nodule
- Pneumonia
- Hernia
- No Finding

### Dataset Imbalance
The dataset is highly imbalanced. Common classes such as "No Finding" have many images, while diseases like Hernia have very few samples.

### Challenges
- Noise in X-ray images
- Annotation errors
- Overlapping diseases
- Low contrast

### Summary
The NIH Chest X-ray dataset is widely used for deep learning-based disease detection. It helps train CNN models for automated diagnosis.

## Dataset 2: HAM10000 Skin Cancer Dataset

### Type of Imaging Data
Dermatoscopic skin lesion images

### Number of Images
10,015 images

### Available Classes / Labels
- Melanoma
- Melanocytic nevi
- Basal cell carcinoma
- Benign keratosis
- Dermatofibroma
- Vascular lesions
- Actinic keratoses

### Dataset Imbalance
The dataset is imbalanced because benign lesion images are much more common than rare cancer types.

### Challenges
- Hair/noise in images
- Similar-looking lesions
- Lighting variations
- Small lesion boundaries

### Summary
HAM10000 is commonly used for skin cancer classification using machine learning and deep learning models.

# Conclusion
Both datasets are important in healthcare AI. Chest X-rays help detect lung diseases, while dermatoscopic images help identify skin cancer. Dataset imbalance and noise remain major challenges in medical imaging.
